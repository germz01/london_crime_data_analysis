### ABOUT THE DATA <a name='about_the_dataset'><a>

I have selected this dataset in order to enrich the main block of information about the criminal activities with some
useful demographics, with the purpose of providing a social context for the events that are represented by the main
dataset. The dataset that I have collected is called **London Borough Profiles**, and is available at
[this address](https://data.london.gov.uk/dataset/london-borough-profiles). The data is set out across 11 themes 
covering most of the key indicators relating to demographic, economic, social and environmental data.
These profiles include data relating to: Population, Households (census), Demographics, Migrant population, 
Ethnicity, Language, Employment, NEET, DWP Benefits (client group), Housing Benefit, Qualifications, Earnings, 
Volunteering, Jobs density, Business Survival, Crime, Fires, House prices, New homes, Tenure, Greenspace, Recycling, 
Carbon Emissions, Cars, Public Transport Accessibility (PTAL), Indices of Multiple Deprivation, GCSE results, 
Children looked after, Children in out-of-work families, Life Expectancy, Teenage conceptions, Happiness levels, 
Political control, and Election turnout. I have selected some of the fields I considered relevant to the analysis'
context.

### LIBRARIES

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

### DATASETS

```python
demographics = pd.read_csv('../data/london-borough-profiles.csv', 
                           encoding="ISO-8859-1")
demo_social = pd.read_csv('../data/demographics_social_2011_2014.csv')
demo_labour = pd.read_csv('../data/demographics_labour_2015.csv',
                          na_values='.')
demo_transports = pd.read_csv('../data/demographics_transports_2011_14_15.csv')
```

```python
demographics.drop(demographics.index[33:38], inplace=True)
```

### SOCIAL PROFILING <a name='social_profiling'></a>

Here I represent some of the most interesting social demographics that are stored in the dataset. This are the
**Life Satisfaction Score**, which measures how people evaluate their life as a whole rather than their current 
feelings, the **Worthwhileness Score**, which measures how people think they are worthwhile or 
valuable, the **Happiness Score**, which measures how people rate their own current lives, and the 
**Anxiety Score**, which measures how people perceive anxiety in their current lives. All the scores have been 
collected from 2011 to 2014, and are reported on a 0 to 10 scale. 

```python
demo_social.describe(include=np.number)
```

#### SUMMARY

As we can see, both the mean value for the **Life Satisfaction Score**, the **Worthwhileness Score** and 
**Happiness Score** are quite high, all being around 7. We can conclude that, in general, the people among the 33
boroughs are quite satisfied with their life. The mean value for the **Anxiety Score** is low, being around 3.

#### PROFILING PER BOROUGH

Here I propose a visualization of the Social Demographics on a per-borough basis.

```python
demo_social.groupby('Borough', sort=True).sum()
```

#### SUMMARY

By observing the Social Demographics on a per-borough basis we can see that the results are coherent with the 
previous general analysis. The values for the **Life Satisfaction Score**, the **Worthwhilleness Score** and the 
**Happiness Score** among the the 33 boroughs are all between 6 and 7, revealing a quite satisfied population. It is 
interesting to note that the highest value for the **Axienty Score**, that is, 5.6, belongs to City of London.

#### VISUALIZATIONS

Here we can see the visualizations for the distributions of each variable for the Social Demographics.

```python
plt.figure(figsize=(10, 10))

for i in range(1, 5):
    plt.subplot(2, 2, i)
    plt.hist(demo_social[demo_social.columns[1:][i - 1]], edgecolor='black', 
             zorder=3, color='#FCD2A8')
    plt.grid(zorder=0)
    plt.title('# of Boroughs per {}'.format(demo_social.columns[1:][i - 1]),
              fontsize=14)
    plt.xlabel(demo_social.columns[1:][i - 1], fontsize=12)
    plt.ylabel('# of Boroughs', fontsize=12)

plt.suptitle('Social Demographics per # of Boroughs', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('../imgs/demographics/social_demographics.pdf', 
            bbox_inches='tight')
```

#### SUMMARY

Observing the distribution for the **Life Satisfaction Score** we can confirm what has been said up to now, that is, 
except for City of London, which a score of 6.6, the other boroughs have a score between 7.0 and 7.6. The 
distribution for the **Worthwhileness Score** reveals three clusters, that is, a cluster composed by boroughs with a
score between 7.5 and 7.9, one composed by boroughs with a score between 7.3 and 7.4 and one composed by only one 
borough, that is, City of London, with a score of 7.1. The distribution for the **Happiness Score** is composed by
two clusters, one composed by boroughs with a score between 7.0 and 7.6 and one composed only by City of London, with
a score of 6.0. Finally, the distribution for the **Anxiety Score** reveals two clusters, one composed by City of 
London, which have a score of 5.6, and the other composed by all the other boroughs, with a score between 2.6 and 4.1.

### LABOUR PROFILING <a name='labour_profiling'></a>

Here I represented some of the labour-related demographics that are stored in the dataset. All the metrics are been
collected in 2015. In my opinion, the most useful ones are **Employment Rate**, **Male** and **Female Employment 
Rate**, **Unemployment Rate** and **Youth Unemployment Rate (18-24)**. Each score is represented in percentage, and
the variables' meanings are quite self-explanatory.

```python
demo_labour.fillna(demo_labour.mean(), inplace=True)
demo_labour.describe()
```

#### SUMMARY

As we can see, the **Employment Rate** is quite high, with a mean value of 72.72% and a maximum value of 79.60%.  We 
can see that there is a sensible difference between the **Male Employmet Rate** and the **Female Employment Rate**,
with the former having a mean value greater than the latter by 12.88 percentage points. The **Unemployment Rate** is 
low, with a mean value of 6.078%, as well as the **Youth Unemployment Rate**, which have a very low mean value of
3.47%.

#### PROFILING PER BOROUGH

Here I propose a visualization of the Labour Demographics on a per-borough basis.

```python
demo_labour.groupby('Borough', sort=True).sum()
```

#### SUMMARY

We can see that, on a per-borough basis the values for the **Employment Rate** are roughly between 65% and 79%, for 
the **Male Employment Rate** they are between 68% and 86%, for the **Female Employment Rate** they are between 56% 
and 71%, for the **Unemployment Rate** they are between 4% and 11% and finally for the **Youth Unemployment Rate**
they are between 2.0% and 5.6%.

#### VISUALIZATIONS

Here we can see the visualizations for the distributions of each variable for the Labour Demographics.

```python
plt.figure(figsize=(12, 14))

for i in range(1, 6):
    plt.subplot(3, 2, i)
    plt.hist(demo_labour[demo_labour.columns[1:][i - 1]], edgecolor='black', 
             zorder=3, color='#EC6778')
    plt.grid(zorder=0)
    plt.title('# of Boroughs per {}'.format(demo_labour.columns[1:][i - 1]),
              fontsize=14)
    plt.xlabel(demo_labour.columns[1:][i - 1], fontsize=12)
    plt.ylabel('# of Boroughs', fontsize=12)

plt.suptitle('Labour Demographics per # of Boroughs', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('../imgs/demographics/labour_demographics.pdf', 
            bbox_inches='tight')
```

#### SUMMARY

For the **Employment Rate**, the **Female Employment Rate** and the **Youth Unemployment Rate** we can see that there
isn't a subdivision in clusters, that is, the variables form a solid distribution. The **Male Employment** variable
have a distribution splitted in two clusters, one composed by only one borough, Westminster, having a score of 
68.6% and the another composed by the other boroughs, with a score between 72.5% and 87.5. The distribution for the
**Unemployment Rate** is splitted in three clusters, one composed by only one borough, Barking and Dagenham, with
a score of 11%, and the other two with, respectively, a score between 4 and 6.9 and a score between 7.1 and 9.2.
